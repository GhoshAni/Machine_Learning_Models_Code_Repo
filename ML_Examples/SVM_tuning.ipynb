```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm , datasets
%matplotlib inline
```

```python
## import iris data
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
def plotSVC(title):
  # create a mesh to plot in
    x_min = X[:, 0].min() - 1
    x_max = X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    h = (x_max / x_min)/100
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h))
    plt.subplot(1, 1, 1)
    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length') 
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.title(title)
    plt.show()
```

```python
## kernels parameter selects the type of hyperplane
kernels = ['linear', 'rbf', 'poly']

for kernel in kernels:
    svc = svm.SVC(kernel = kernel).fit(X,y)
    plotSVC('kernel=' + str(kernel))
    
```

```python
## GAMMA gamma is a parameter for non linear hyperplanes, the higher the gamma value, higher the overfitting
gammas = [0.1, 0.5, 0.7, 1,5, 10, 100]
for gamma in gammas:
    svc = svm.SVC(kernel = 'rbf', gamma = gamma).fit(X ,y)
    plotSVC('gamma =' + str(gamma))

## we can see that increasing gamma leads to overfitting  as the classifier tries to perfecrly fit the training data
```

```python
## parameter C: parameter C is associated with margin
## regularization parameter
## large value of parameter C => small margin => lower missclassification
## small value of parameter C = higher margin => higher misclassfication
## tradeoff btw smooth decision boundary and claasifying the training poins corectly
cs = [0.1, 1 , 10, 100, 1000]
for c in cs:
    svc = svm.SVC(kernel = 'rbf', C = c).fit(X,y)
    plotSVC('C=' + str(c))
```

```python
degrees = [0,1,2,3,4,5,6]
for degree in degrees:
    svc = svm.SVC(kernel = 'rbf', degree = degree).fit(X,y)
    plotSVC('degree='+ str(degree))
```
